**4. Plot of Piecewise functions**

**Example**

$$f(x) = \begin{cases}1+x^2; x\in(0.1)\\ 1+x; x\in [-1,0]\end{cases}$$

```python
# First define a piecewsie functions
fp = piecewise([((0,1),1+x^2),([-1,0],1+x)])
#Plot of piecewise function
plot(fp,figsize = 3)
```

**Example**



$$
f(x) =\begin{cases} -1; -1<x<0.5\\1+x^2;0.5<x<1.5\\2x^3-3;1.5<x<2.5 \end{cases}
$$



```python
#Define a piecewise functions
f1(x) = -1
f2(x) = 1-x^2
f3(x) = 2*x^3-3
# put into piecewise function
pf = piecewise([[(-1,0.5),f1],[(0.5,1.5),f2],[(1.5,2.5),f3]])
# See the plot
pf.plot((x,-1,2.5),figsize = 4)
```

**Plot of the graph of the function given by parametric coordinates$$x = tsin(1-t^2), y = tcos(t^2):-2\pi \leq t \leq 2\pi$$

```python
# See the plot
var('t')
p = parametric_plot([t*sin(1-t^2),t*cos(t^2)],(t,-2*pi,2*pi),color = 'red',thickness = 1.2)
# Add the ttext
txt = text('Nce PLot',(-4,4))
show(p+txt,figsize = 4)
```
